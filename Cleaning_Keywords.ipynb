```python
#Removing repeat words from each row
def word_replace(test_list):
  res = []
  [res.append(x) for x in test_list if x not in res]
  return res
```

```python
import pandas as pd
import keras
from keras import layers
from keras import models
from tensorflow.keras import optimizers
import numpy as np
from keras.layers import Dense, Activation, Flatten, Dropout, BatchNormalization
```

```python
df = pd.read_excel('/content/drive/MyDrive/onehotcoded.xlsx')
```

```python
df['final_keywords'] = df['final_keywords'].apply(eval)
```

```python
first = df.sample(n = 9299, replace = False)
first.drop(columns = ['Unnamed: 0','final_keywords'], axis =1, inplace = True)
first.loc[:,'Record ID'] = first['Record ID'].astype(str) + '.jpg'
```

```python
train = first[:6000]
validation = first[6000:7500]
test = first[7000:]
```

```python
from keras_preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(
    rescale=1./255.,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
test_datagen=ImageDataGenerator(rescale=1./255.)
```

```python
columns = train.columns[1:].tolist()
```

```python
train_generator=datagen.flow_from_dataframe(
dataframe=train,
directory="/content/drive/MyDrive/Image Classification Data/images",
x_col='Record ID',
y_col=columns,
batch_size=32,
seed=42,
shuffle=True,
class_mode="raw",
target_size=(256,256))
```

```python
validation_generator=test_datagen.flow_from_dataframe(
dataframe=validation,
directory="/content/drive/MyDrive/Image Classification Data/images",
x_col="Record ID",
y_col=columns,
batch_size=32,
seed=42,
shuffle=True,
class_mode="raw",
target_size=(256,256))
```

```python
test_generator=test_datagen.flow_from_dataframe(
dataframe=test,
directory="/content/drive/MyDrive/Image Classification Data/images",
x_col="Record ID",
batch_size=1,
seed=42,
shuffle=False,
class_mode=None,
target_size=(256,256))
```

```python
from keras.applications.vgg16 import VGG16 
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(256, 256, 3))
conv_base.summary()
```

```python
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(141, activation='sigmoid'))
```

```python
conv_base.trainable = True
set_trainable = False
for layer in conv_base.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=validation_generator.n//validation_generator.batch_size
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size
```

```python
model.compile(loss='binary_crossentropy',
                      optimizer=optimizers.RMSprop(learning_rate=0.0001, decay=1e-6),
                      metrics=[keras.metrics.Precision(top_k=5)])
```

```python
history = model.fit(train_generator,
            steps_per_epoch=STEP_SIZE_TRAIN,
            validation_data=validation_generator,
            validation_steps=STEP_SIZE_VALID,
            epochs = 30
)
```
